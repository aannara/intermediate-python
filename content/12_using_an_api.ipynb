## Fetching data from HSDES

https://hsdes-api.intel.com/rest/doc/

https://wiki.ith.intel.com/display/HSDESWIKI/Python

```python
import requests
from requests_kerberos import HTTPKerberosAuth
import urllib3
 
 
headers = { 'Content-type': 'application/json' }
# Replace the ID here with some article ID
url = 'https://hsdes-api.intel.com/rest/article/2202652508'
response = requests.get(url, verify=False, auth=HTTPKerberosAuth(), headers = headers)
pprint(response.json())
```
